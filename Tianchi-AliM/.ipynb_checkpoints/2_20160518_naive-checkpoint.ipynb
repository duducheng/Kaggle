```python
user_play = user[user.action_type==1]
user_download = user[user.action_type==2]
user_like = user[user.action_type==3]
```

```python
def reindex_df(df):
    df.index = map(lambda t:t[0],df.index)
    df.index.name='song_id'
    return df
```

```python
clean_play = reindex_df(user_play[['song_id','Ds']].groupby('song_id').apply(lambda g: g.groupby('Ds').count().T).fillna(0))
```

```python
clean_download = reindex_df(user_download[['song_id','Ds']].groupby('song_id').apply(lambda g: g.groupby('Ds').count().T).fillna(0))
```

```python
clean_like = reindex_df(user_like[['song_id','Ds']].groupby('song_id').apply(lambda g: g.groupby('Ds').count().T).fillna(0))
```

```python
from dateutil.parser import parse as dateutil_parser
```

```python
def recolumn_df(df):
    df.columns = df.columns.map(lambda x: dateutil_parser(str(x)))
    return df
```

```python
clean_play = recolumn_df(clean_play)
clean_download = recolumn_df(clean_download)
clean_like = recolumn_df(clean_like)
```

```python
clean_play = pd.read_csv(DATA_PATH+'naive/play.csv',index_col=0)
clean_download = pd.read_csv(DATA_PATH+'naive/download.csv',index_col=0)
clean_like = pd.read_csv(DATA_PATH+'naive/like.csv',index_col=0)
```

```python
song.index = song.song_id
all_songs = song.loc[:,['song_init_plays']]
all_songs['howold'] = (song['publish_time'].map(lambda x: dateutil_parser(str(x))) - dateutil_parser('20150301')).map(lambda x: x.days)
lang = pd.get_dummies(song['Language'],'lang')
gender = pd.get_dummies(song['Gender'],'gender')
dummy_features = lang.columns.union(gender.columns)
all_songs = pd.concat([all_songs,lang,gender],axis=1)
```

```python
train_play = clean_play.iloc[:,:-60].rename(columns=lambda x: str(x)+'_play')
train_download = clean_download.iloc[:,:-60].rename(columns=lambda x: str(x)+'_download')
train_like = clean_like.iloc[:,:-60].rename(columns=lambda x: str(x)+'_like')

Xtrain = pd.concat([all_songs,train_play,train_download,train_like],axis=1).fillna(0)
Ytrain = pd.concat([pd.DataFrame(index=all_songs.index),clean_play.iloc[:,-60:]],axis=1).fillna(0)
```

```python
print Xtrain.shape
print Ytrain.shape
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.cross_validation import cross_val_score
```

```python
rfr = RandomForestRegressor(100)
```

```python
val_scores = cross_val_score(rfr, Xtrain, Ytrain)
```
