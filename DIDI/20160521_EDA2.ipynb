```python
def pclass(p):
    class_set = range(5,20,5)+range(20,100,10)+range(100,501,100)
    idx_set = [p>cls for cls in class_set]
    return idx_set.index(False) if sum(idx_set)!=len(idx_set) else len(idx_set)
```

```python
test_order['timeslot'] = test_order['Time'].map(lambda x: (x.hour*60+x.minute)/10+1)
test_order['datetimeslot'] = test_order['Time'].map(lambda x: x.year*10000+x.month*100+x.day)*1000+test_order['timeslot']
test_order['pclass'] = test_order['Price'].map(lambda x: pclass(x))
test_order = pd.concat([test_order,pd.get_dummies(test_order['pclass'],'pclass').applymap(lambda x: {1.0: True, 0.0: False}[x])],axis=1)
```

```python
def map_group(group):
    res = pd.Series()
    res['request'] = group['request'].count()
    res['answer'] = group['request'].sum()
    res['price_avg'] = group['Price'].mean()
    pclass_cols = filter(lambda x: x[:7]=='pclass_',group.columns)
    pclass_values = group[pclass_cols].sum()
    res = pd.concat([res,pclass_values])
    return res
```

```python
order_sample_group = order_sample.groupby('datetimeslot').apply(lambda g: map_group(g))
```
