```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from dateutil.parser import parse as date_parser

from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from sklearn.grid_search import GridSearchCV
from sklearn.preprocessing import MinMaxScaler
class Scaler(MinMaxScaler):
    def transform(self,X):
        return pd.DataFrame(super(Scaler,self).transform(X),index=X.index,columns=X.columns)

%matplotlib inline
```

```python
DETROIT_LAT = (42.252, 42.452)
DETROIT_LNG = (-83.295, -82.895)
```

```python
clean_permit = pd.read_csv('clean/permit.csv',index_col=0)
clean_violation = pd.read_csv('clean/violation.csv',index_col=0)
clean_crime = pd.read_csv('clean/crime.csv',index_col=0)
clean_311 = pd.read_csv('clean/311.csv',index_col=0)
```

```python
# clean_permit
```

```python
def clean(df):
    df = df.dropna()
    df = df[(df['lat']>=DETROIT_LAT[0]) & (df['lat']<=DETROIT_LAT[1])]
    df = df[(df['lng']>=DETROIT_LNG[0]) & (df['lng']<=DETROIT_LNG[1])]
    if hasattr(df,"turst"):
        df = df[df.turst]
        df = df.drop('turst',axis=1)
    df['date'] = df['date'].map(date_parser)
    return df
```

```python
clean_permit, clean_violation, clean_crime, clean_311 = map(clean,[clean_permit, clean_violation, clean_crime, clean_311])
```

```python
# tmp = clean_permit, clean_violation, clean_crime, clean_311
```

```python
addr = clean_violation[['lat','lng']]
scaler = Scaler().fit(addr)
```

```python
def scale(df):
    df[['lat','lng']] = scaler.transform(df[['lat','lng']])
    return df
clean_permit, clean_violation, clean_crime, clean_311 = map(scale,[clean_permit, clean_violation, clean_crime, clean_311])
```
