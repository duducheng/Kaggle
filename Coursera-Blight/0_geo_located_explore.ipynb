```python
raw_violations = pd.read_csv(DATA_PATH + 'detroit-blight-violations.csv', low_memory=False)
raw_violations.head()
```

```python
lat_lng = raw_violations['ViolationAddress'].map(lambda x: x.split('\n')[-1])
```

```python
lat = lat_lng.map(lambda x: int(1000*float(x.split(',')[0][1:]))/1000.0)
lng = lat_lng.map(lambda x: int(1000*float(x.split(',')[1][:-1]))/1000.0)
```

```python
print(lat.isnull().sum())
print(lng.isnull().sum())
```

```python
pd.concat([lat,lng])[2]
```

```python
addr = pd.DataFrame()
addr['lat'] = lat
addr['lng'] = lng
addr['city'] = 'Detroit'
addr.index.name = 'id'
# addr.to_csv('addr4Tableau.csv')
```
